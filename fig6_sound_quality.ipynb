```python
#GRAFICA DEL SNR Y MI PARA DOS SEALES DISTINTAS

samplerate, data1  = read(wav_directory +"grazing_good.wav")
length = data1.shape[0] / samplerate
time = np.linspace(0., length, data1.shape[0])
grazing_good = data1.astype(np.float32, order='C') / 32768.0
del data1
_, data2  = read(wav_directory +"grazing_bad.wav")
grazing_bad = data2.astype(np.float32, order='C') / 32768.0
del data2
_, data3  = read(wav_directory +"rumination_good.wav")

# Create main container with size of 6x5
fig = plt.figure(figsize=(16, 8))
plt.rcParams['text.usetex'] = True
font = {'family':'serif', 'serif': ['computer modern roman']}
plt.rc('font',**font)

plt.subplots_adjust(bottom = 0., left = 0, top = 1., right = 1, hspace=0.8)
#plt.suptitle('Grazing', y= 1.1, fontsize=16)

sub1 = fig.add_subplot(3,2,1) 
sub1.plot(time, grazing_good, color = 'black',linewidth=0.5)
sub1.set_xlim(0, 20)
sub1.set_ylim(-1, 1)
sub1.set_ylabel("Amplitude relative", labelpad = 5, fontsize=18)
sub1.set_xlabel("Time [s]\n\n(a)", labelpad = 5, fontsize=18)
sub1.title.set_text('Grazing')
sub1.title.set_fontsize(18)
sub1.legend(['$MI_{JM} = 0.73$ \n $SNR = 10.59$'],fontsize=14,
            handletextpad=0.0, handlelength=0,framealpha=1,loc='upper right')

# Create second axes, the bottom plot (original signal)
sub2 = fig.add_subplot(3,2,2) 
sub2.plot(time, grazing_bad, color = 'black',linewidth=0.5)
sub2.set_xlim(0, 20)
sub2.set_ylim(-1, 1)
sub2.set_ylabel("Amplitude relative", labelpad = 5, fontsize=18)
sub2.set_xlabel("Time [s]\n\n(b)", labelpad = 5, fontsize=18)
sub2.title.set_text('Grazing')
sub2.title.set_fontsize(18)
sub2.legend(['$MI_{JM} = 0.31$ \n $SNR = 7.97$'],fontsize=14,
            handletextpad=0.0, handlelength=0,framealpha=1,loc='upper right')

plt.savefig(fig_directory +'fig6.jpg', dpi = 600,transparent=True,bbox_inches='tight')
plt.savefig(fig_directory +'fig6.pdf', dpi = 600,transparent=True,bbox_inches='tight')
```
